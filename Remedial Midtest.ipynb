```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan histogram grayscale
def histogram_grayscale(image, title):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Mengonversi gambar menjadi grayscale
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram Grayscale - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.hist(gray.ravel(), 256, [0, 256], color='gray', alpha=0.7)
    plt.grid(True)
    plt.show()

# Fungsi utama
def main():
    # Path gambar (masukkan path gambar yang akan digunakan)
    file_path = "Doyong1.jpeg"  # Ganti dengan path gambar Anda

    # Membaca gambar
    image_path = "view.jpg"
    image = cv2.imread(file_path)

    if image is None:
        print("Gambar tidak ditemukan! Periksa kembali path gambar.")
        return

    # Menampilkan histogram grayscale untuk gambar
    histogram_grayscale(image, "Gambar")

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan histogram channel biru
def histogram_blue(image, title):
    blue_channel = image[:, :, 0]  # Ambil channel biru
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram Blue - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.hist(blue_channel.ravel(), 256, [0, 256], color='blue', alpha=0.7)
    plt.grid(True)
    plt.show()

# Fungsi untuk menampilkan histogram channel hijau
def histogram_green(image, title):
    green_channel = image[:, :, 1]  # Ambil channel hijau
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram Green - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.hist(green_channel.ravel(), 256, [0, 256], color='green', alpha=0.7)
    plt.grid(True)
    plt.show()

# Fungsi untuk menampilkan histogram channel merah
def histogram_red(image, title):
    red_channel = image[:, :, 2]  # Ambil channel merah
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram Red - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")
    plt.hist(red_channel.ravel(), 256, [0, 256], color='red', alpha=0.7)
    plt.grid(True)
    plt.show()


# Fungsi untuk menampilkan histogram RGB secara gabungan
def histogram_rgb_combined(image, title):
    plt.figure(figsize=(8, 6))
    plt.title(f"Histogram RGB Gabungan - {title}")
    plt.xlabel("Intensitas Pixel")
    plt.ylabel("Jumlah Pixel")

    # Mengambil histogram untuk setiap channel
    hist_b = cv2.calcHist([image], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([image], [1], None, [256], [0, 256])
    hist_r = cv2.calcHist([image], [2], None, [256], [0, 256])

    # Menggabungkan ketiga channel dalam satu plot
    plt.fill_between(range(256), hist_b.flatten(), color='blue', alpha=0.3, label='Blue Channel')
    plt.fill_between(range(256), hist_g.flatten(), color='green', alpha=0.3, label='Green Channel')
    plt.fill_between(range(256), hist_r.flatten(), color='red', alpha=0.3, label='Red Channel')

    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

# Fungsi utama
def main():
    # Path gambar kedua (gambar berwarna)
    file_path2 = "Doyong2.jpeg"  # Ganti dengan path gambar kedua (warna)

    # Membaca gambar kedua
    image2 = cv2.imread(file_path2)

    if image2 is None:
        print("Gambar kedua tidak ditemukan!")
        return

    # Menampilkan histogram untuk setiap channel secara terpisah
    print("\nHistogram untuk channel Biru:")
    histogram_blue(image2, "Gambar 2")

    print("\nHistogram untuk channel Hijau:")
    histogram_green(image2, "Gambar 2")

    print("\nHistogram untuk channel Merah:")
    histogram_red(image2, "Gambar 2")

    # Menampilkan histogram gabungan RGB untuk gambar kedua
    print("\nHistogram gabungan RGB untuk gambar kedua:")
    histogram_rgb_combined(image2, "Gambar 2")

if __name__ == "__main__":
    main()
```
